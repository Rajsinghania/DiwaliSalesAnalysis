**Gender Distribution**
**Methond 1**

```python
plt.figure(figsize=(5,5))
df['Gender'].value_counts().plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
# Add counts on top of each bar
for index, value in enumerate(df['Gender'].value_counts()):
    plt.text(index, value, str(value), ha='center', va='bottom')  # Adjust 'va' as needed

plt.show()
```

**Gender Distribution**
**Methond 2**

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(x='Gender', data=df, color='skyblue')
# Add labels on top of each bar
ax.bar_label(ax.containers[0])
plt.title('Gender Distribution')
plt.show()
```

**Sales_Gender_Wise**

```python
sales_Gender=df.groupby(['Gender'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(5,5))
ax = sns.barplot(x='Gender', y='Amount', data=sales_Gender)
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

**Marital_Status_Distribution**

```python
#Marital_status
ax = sns.countplot (x= 'Marital_Status', data=df, hue='Gender')
#ax.bar_label(ax.containers[0])
for container in ax.containers:
    ax.bar_label(container)
plt.title('Marital_Status Distribution')
plt.show()
```

**Sales_Marital_Status**

```python
Sales_Marital_Status = df.groupby(['Marital_Status', 'Gender'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending= False)
ax = sns.barplot(x='Marital_Status', y='Amount' , data = Sales_Marital_Status, hue = 'Gender')
for container in ax.containers:
    ax.bar_label(container)
plt.ylim(0, Sales_Marital_Status['Amount'].max() * 1.1)
plt.show()
print(Sales_Marital_Status)
```

**Age_Distribution**

```python
#Age Distribution
ax = sns.countplot(x= 'Age Group', data=df, hue = 'Gender')
#ax.bar_label(ax.containers[0])
for container in ax.containers:
    ax.bar_label(container)
plt.title('Age Group')
plt.show()
```

**Sales_Age_Group**

```python
sales_Age_group = df.groupby(['Age Group', 'Gender'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data= sales_Age_group, hue='Gender')
plt.show()
#print(sales_Age_group)
```

**State_wise_Sales & Order_Distribution**

Sales_Distribution

```python
sales_Statewise= df.groupby(['State'], as_index=False ) ['Amount'].sum().sort_values(by='Amount', ascending= False).head(10)
plt.figure(figsize=(15, 6))
sns.barplot(x='State', y='Amount', data = sales_Statewise)
plt.show()
print(sales_Statewise)
```

Order Distribution

```python
State_order = df.groupby(['State'], as_index=False) ['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
plt.figure(figsize=(15,6))
sns.barplot(x='State', y='Orders', data = State_order)
plt.show()
```
